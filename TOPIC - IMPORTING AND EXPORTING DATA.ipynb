```python
import pandas as pd
```

```python
columns = ['Date', 'Store_Number', 'Transaction_Count']
trns = pd.read_csv('transactions.csv',
                   parse_dates = ['Date'],
                   header = 0,
                   names = columns,
                   dtype={'Store_Number' : 'int8', 'Transaction_Count' : 'int16'}).assign(
                        pct_to_target = trns['Transaction_Count'] / 2500,
                        met_target = (trns['Transaction_Count'] / 2500) >= 1,
                        bonus_payable = ((trns['Transaction_Count'] / 2500) >= 1) * 100,
                        month = trns['Date'].dt.month,
                        day_of_week = trns['Date'].dt.dayofweek,  
                  ).astype({ 'pct_to_target' : 'float32','month' : 'int8','day_of_week' : 'int8'})
                  
trns
```

```python
trns.info(memory_usage='deep')
```

```python
#ASSIGNMENT EXPORTING DATA
with pd.ExcelWriter('Data For Chandler.xlsx') as writer:
    trns.loc[trns['Date'].dt.year == 2013].to_excel(writer, sheet_name='2013')
    trns.loc[trns['Date'].dt.year == 2014].to_excel(writer, sheet_name='2014')
    trns.loc[trns['Date'].dt.year == 2015].to_excel(writer, sheet_name='2015')
    trns.loc[trns['Date'].dt.year == 2016].to_excel(writer, sheet_name='2016')
    trns.loc[trns['Date'].dt.year == 2017].to_excel(writer, sheet_name='2017')
```

```python
#or we can use this method instead
with pd.ExcelWriter('Data For Chandler2.xlsx') as writer:
    for year in range (2013, 2018):
        trns.loc[trns['Date'].dt.year == year].to_excel(writer, sheet_name=str(year))
        
```
