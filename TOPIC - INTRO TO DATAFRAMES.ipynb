```python
#ASSIGNMENT COLUMN CREATION
import numpy as np
import pandas as pd
transactions
```

```python
#creating target column
transactions['pct_to_target'] = transactions['transaction_count'] / 2500
transactions['met_target'] = transactions['pct_to_target'] >= 1
transactions['bonus_payable'] = (transactions['met_target'] >= 1) * 100
transactions
```

```python
#creating date column
transactions['date'] = transactions['date'].astype("datetime64[ns]")
transactions['month'] = transactions['date'].dt.month
transactions['day_of_week'] = transactions['date'].dt.dayofweek
transactions
```

```python
#ASSIGNMENT NUMPY SELECT
import numpy as np
import pandas as pd
transactions
```

```python
#create seasonal_bonus column that applies to: all days in december, sundays in may, mondays in July
#first create the condition
conditions = [
    (transactions['month'] == 12), 
    (transactions['month'] == 5) & (transactions['day_of_week'] == 6),
    (transactions['month'] == 7) & (transactions['day_of_week'] == 0),
]

#create the choices/options 
choices = ['Holiday Bonus', 'Corporate Month', 'Summer Special']

#apply the condition and choices using the select method
transactions['seasonal_bonus'] = np.select(conditions, choices, default='None')

transactions
```

```python
#check the info of the df
transactions['seasonal_bonus'].value_counts()
```

```python
#calculate total bonus at 100$ a day
transactions['seasonal_bonus'].isin(['Holiday Bonus', 'Summer Special', 'Corporate Month']).sum() * 100
```

```python
#ASSIGNMENT MAP AND ASSIGN
#drop the column first with the following name:
transactions.drop(['pct_to_target','met_target','bonus_payable','month','day_of_week','seasonal_bonus'], axis = 1)
```

```python
#add all of thos column again using assign method
transactions.assign(
    pct_to_target = transactions['transaction_count'] / 2500,
    met_target = transactions['pct_to_target'] >= 1,
    bonus_payable = (transactions['met_target'] >= 1) * 100,
    month = transactions['date'].dt.month,
    day_of_week = transactions['date'].dt.dayofweek,   
    seasonal_bonus = np.select(conditions, choices, default='None')
)
```

```python
#sum the seasonal bonus again
transactions['seasonal_bonus'].isin(['Holiday Bonus', 'Summer Special', 'Corporate Month']).sum() * 100
```

```python
#or, we can sum it using this method
transactions['seasonal_bonus'].value_counts().iloc[1:].sum() * 100
```
