```python
import numpy as np
import pandas as pd
```

```python
transactions = pd.read_csv('transactions.csv', parse_dates=['date'] )
transactions
```

```python
#ASSIGNMENT GROUPBY
#group the dataframe based on 10 stores with the most transaction 
transactions.groupby('store_nbr')[['transactions']].sum().sort_values(['transactions'], ascending=False).head(10)
```

```python
#ASSIGNMENT GROUPBY MULTIPLE COLUMNS
#total transaction by store and month, month is asc, transaction is desc
#first, extract month from date
transactions['month'] = transactions['date'].dt.month
transactions
```

```python
#total transaction by store and month, month is asc, transaction is desc

(transactions.groupby([ 'store_nbr','month'])[['transactions']]
.sum()
.sort_values(['month','transactions'], ascending=[True,False])
)
```

```python
#ASSIGNMENT MULTI INDEX DATAFRAMES
transactions
```

```python
#create first aggregation table
grouped = (
    transactions.groupby(['store_nbr','month'])
    .agg({'transactions' : ['sum','mean']})
    .sort_values(by=['month', ('transactions','sum')], ascending=[True,False])
)

grouped
```

```python
#grab store 3 month 1
grouped.loc[(3,1)]
```

```python
#select store 3 month 1 AND the mean 
grouped.loc[(3,1), [('transactions', 'mean')]]    #add to layer on loc method, for row and column

#berarti, 1 [] itu adalah 1 column, kalo mau grab 2 column yg berbeda pake (), okok
```

```python
grouped
```

```python
#create integer-based index, drop the transactions title
grouped.droplevel(0, axis=1).reset_index()
```

```python
#ASSIGNMENT THE AGG METHOD
#create again the transaction df from previous topic, named it transactions2

transactions2 = transactions.assign(
    pct_to_target = transactions['transactions'] / 2500,
    met_target = (transactions['transactions'] / 2500) >= 1,
    bonus_payable = ((transactions['transactions'] / 2500) >= 1) * 100,
    month = transactions['date'].dt.month,
    day_of_week = transactions['date'].dt.dayofweek,   
)
transactions2
```

```python
#create a table where avg number of days each store hit the target
transactions2.groupby('store_nbr').agg({'met_target':'mean'})
```

```python
#calculate total bonus payable and sort from the highest bonus to lowest, by stor number
(transactions2.groupby('store_nbr')
    .agg({'met_target': 'mean',
         'bonus_payable' : 'sum'})
).sort_values('bonus_payable', ascending=False).head()
```

```python
#calculate total bonus payable and sort from the highest bonus to lowest, by month
(transactions2.groupby('month')
    .agg({'met_target': 'mean',
         'bonus_payable' : 'sum'})
).sort_values('bonus_payable', ascending=False).head()
```

```python
#calculate total bonus payable and sort from the highest bonus to lowest, by weekday
(transactions2.groupby('day_of_week')
    .agg({'met_target': 'mean',
         'bonus_payable' : 'sum'})
).sort_values('bonus_payable', ascending=False).head()
```

```python
#ASSIGNMENT TRANSFORMING A DATAFRAME
transactions2.assign(
    store_avg_trans = transactions2.groupby(['store_nbr','day_of_week'])['transactions'].transform('mean'),
    trans_vs_avg = lambda x: x['transactions'] - x['store_avg_trans']
).head()
```

```python
#ASSIGNMENT PIVOT AND MELT
transactions2
```

```python
transactions2.info()
```

```python
#summarize store number 1 to 11 with values of sum of bonus payable using pivot, after that, use cmaps, also change 0 to nan
cmap_transactions = transactions2.pivot_table(index='store_nbr',
                          columns='day_of_week',
                          aggfunc='sum',
                          values='bonus_payable'
).replace(0, np.nan).iloc[:11].style.background_gradient(cmap='RdYlGn', axis=1)

cmap_transactions
```

```python
#or another method for the NAN 
(transactions2.loc[transactions2['bonus_payable'] != 0]
    .pivot_table(index='store_nbr', columns='day_of_week', aggfunc='sum', values='bonus_payable')
).loc[:11].style.background_gradient(cmap='RdYlGn', axis=1)
```

```python
#drop styler first because we cant unpivot it, it will show error because it considered the melt method as styler
cmap_transactions = transactions2.pivot_table(index='store_nbr',
                          columns='day_of_week',
                          aggfunc='sum',
                          values='bonus_payable'
).replace(0, np.nan).iloc[:11]

cmap_transactions
```

```python
#reset index first, then melt
(cmap_transactions.reset_index()
    .melt(id_vars='store_nbr', #value_vars='day_of_week', 
          value_name='bonus_payable'
    )
)
```
